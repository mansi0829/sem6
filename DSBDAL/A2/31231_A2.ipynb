```python
#As all the FIVE scores have median ~ mode
#The data points which fall below mean-3*(std.dev.) or above mean+3*(std.dev.) are outliers.
#Z-Score Treatment
def trim(col:pd.core.series.Series):
    upperLimit = col.mean() + 5 * col.std()
    lowerLimit = col.mean() - 5 * col.std()
    newCol = col[(col < upperLimit) & (col > lowerLimit)]
    return newCol
```

```python
df2["DBMS"]=trim(df2["DBMS"])
df2["TOC"]=trim(df2["TOC"])
df2["CNS"]=trim(df2["CNS"])
df2["SPOS"]=trim(df2["SPOS"])
df2["DS"]=trim(df2["DS"])
```
